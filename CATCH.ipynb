>Nevertheless, the results of this study can be broadly taken to support the position that focused police enforcement efforts can be effective in reducing crime at hot spots.

## Methodology

The population data will amount to 10 integers, 9 values UCR reports, and the 2019 value from the police department's operational summary. The values will be entered manually. Likewise, the police employment data, also to be entered manually, will be retrieved from UCR reports and the 2019 operational summary.

The crime data are archived by month in publically accessible excel files. Each of these monthly files will be aggregated by year and only a limited amount of information will be extracted for analysis. The only fields intended for use are the Offense Type, the year and month portion of the date, and the count of offenses per occurence.

The analysis strategy is to take place over phases. The first phase will process each month of each year individually by grouping all crimes by their type, and their counts will be summed. Each crime will also be given a binary indicator as either violent or non-violent. From here, all crimes will again be aggregated by their violent-indicator in a second phase. Violent crime will be assessed as an independent group from another group that considers overall crime. Once, for instance, all of the numbers for 2018 are aggregated, the month column will be discarded. In the final phase, the resulting 11 data points for violent crime and the 11 for overall crime will be plotted over a time series by their year. Since the research questions center around crime trends over time, a time series is the ideal approach to communicate the findings. 

The manually recorded population and police employment numbers will also be presented as annual numbers. Each will be shown in individual trends, but an additional measure will combine the two into a ratio. The ratio will be presented as the number of employed officers over every 1,000 citizens. For example, 2,000 officers for a population of 1,500,000 people will be recorded as 1.33. The math follows:

>2000/(1500000/1000) = 2000/1500 = 1.33

This ratio aligns with how the Houston Police Department records their ratio and also provide a single number to track over time. Hopefully, this will make the information simpler to digest.

This following block defines the full gamut of functions used in the dask parallel processing of the HPD archived data. The use of the functions is demonstrated in **Appendix B**.

```python
@delayed
def PrepMonth(m, yr, accountable = False):
    xlFile = 'crime' + yr + '.xlsx'
    sheet = m+yr[2:]
    
    if accountable:
        colFilter = pd.Series(['OffenseType', 'OffenseCount', 'OccurrenceDate'])
        df = pd.read_excel(archive_dir /xlFile, sheet_name=sheet)[colFilter]
        
        # Adjust date column to only store the year
        if df.OccurrenceDate.any():
            df.OccurrenceDate = pd.to_datetime(df.OccurrenceDate).dt.year
            
        df = df.rename(columns={'OccurrenceDate':'Year'})
    else:
        colFilter = pd.Series(['OffenseType', 'OffenseCount'])
        df = pd.read_excel(archive_dir /xlFile, sheet_name=sheet)[colFilter]
        
    return df

@delayed
def MonthPairs(pairArgs):
    #appends months together
    df = pairArgs[0][0]
    return df.append(pairArgs[0][1])

@delayed
def ProcessYear(yr, args, accountable = False):
    # Form 6 month-pairs
    pairs = []
    
    clear_output()
    print(str("Processing {}...").format(yr))
    
    # Process months in pairs
    for i,j in zip(np.arange(0,12,2), np.arange(1,13,2)):
        pairs.append([(args[i], args[j])])
    
    DualMonths = [MonthPairs(k).compute() for k in pairs]
    
    # Form 3 month-groups
    pairs = []
    
    for i,j in zip(np.arange(0,6,2), np.arange(1,7,2)):
        pairs.append([(args[i], args[j])])
        
    GroupMonths = [MonthPairs(k).compute() for k in pairs]
    
    # Build out one table for the entire year
    FullYear = GroupMonths[0]
    FullYear = FullYear.append(GroupMonths[1]).append(GroupMonths[2]).reset_index()
    
    # drop index column
    FullYear = FullYear.drop(['index'], axis=1)
    
    # No need to append the year column in this case
    if accountable:
        return FullYear
    
    # Append Year column
    yearCol = pd.DataFrame([yr]*len(FullYear), columns=['Year'])
    
    return pd.concat((FullYear, yearCol), axis=1)
```

The following functions are all used to aggregate specific crime data into particular formats.

```python
# Aggregate all crime by year
def AggCrime(crimeData, suffix=''):
    df = crimeData[0]
    for i in crimeData[1:]:
        df = df.append(i)
    
    # Clean up OffenseType data
    df.OffenseType = [str(i).rstrip().replace('AutoTheft','Auto Theft').replace('Motor vehicle theft','Auto Theft'). \
                      replace('Burglary, Breaking and Entering','Burglary').replace('Forcible rape','Rape'). \
                      replace('Murder, non-negligent','Murder').replace('All other larceny','Theft'). \
                      replace('From coin-operated machine or device','Theft').replace('Pocket-picking','Theft'). \
                      replace('Purse-snatching','Theft').replace('Shoplifting','Theft'). \
                      replace('Theft from building','Theft').replace('Theft from motor vehicle','Theft'). \
                      replace('Theft of motor vehicle parts or accessory','Theft') for i in df.OffenseType]
    
    #Aggregate all crime
    df = df[df['OffenseType'].isin(['Aggravated Assault','Auto Theft','Burglary','Murder','Rape','Robbery','Theft'])]
    df = df.drop('OffenseType', axis=1)
    df = df.groupby(['Year']).sum().reset_index()
    df.Year = [int(i) for i in df.Year]
    df = df.rename({'OffenseCount':'HPD_All' + suffix},axis=1)
    
    return df

# Aggregate violent crime by year
def AggViolentCrime(crimeData, suffix=''):
    df = crimeData[0]
    for i in crimeData[1:]:
        df = df.append(i)
    
    # Clean up OffenseType data
    df.OffenseType = [str(i).rstrip().replace('Forcible rape','Rape').replace('Murder, non-negligent','Murder') \
                      for i in df.OffenseType]
    
    #Aggregate only violent crime
    df = df[df['OffenseType'].isin(['Aggravated Assault','Murder','Rape','Robbery'])]
    df = df.drop('OffenseType', axis=1)
    df = df.groupby(['Year']).sum().reset_index()
    df.Year = [int(i) for i in df.Year]
    df = df.rename({'OffenseCount':'HPD_Violent' + suffix},axis=1)
    
    return df

# Aggregate FBI data by year
def AggFBI(byType=False):
    inFile =  pd.ExcelFile('CrimeAnalysis.xlsx')
    years = list(map(str,np.arange(2011,2019,1)))
    colFilter = pd.Series(['Year','OffenseType','FBI_OffenseCount'])
    
    df = inFile.parse('crime2010')[colFilter]
    for i in years:
        df = pd.concat((df,inFile.parse('crime'+i)[colFilter]))
    
    if byType:
        dfTypes = df[df.OffenseType.isin(['Assault','Murder','Rape','Robbery'])]
        return dfTypes
    
    else:
        dfTotal = df[df.OffenseType == 'Total']
        dfTotal = dfTotal.drop('OffenseType', axis=1)
        dfTotal = dfTotal.rename({'FBI_OffenseCount':'FBI_All'},axis=1)
        
        dfViolentTotal = df[df.OffenseType == 'Violent Total']
        dfViolentTotal = dfViolentTotal.drop('OffenseType', axis=1)
        dfViolentTotal = dfViolentTotal.rename({'FBI_OffenseCount':'FBI_Violent'},axis=1)
        
        return dfTotal, dfViolentTotal

# Retrieve officer count, population, and officer:population ratio by year
def getOfficerData():
    return DF(pd.read_excel('CrimeAnalysis.xlsx', sheet_name='officerRatio'))

# Function to merge all data into one DataFrame
def MergeData(tables):
    mergeDF = tables[0]
    for df in tables[1:]:
        mergeDF = mergeDF.merge(df, on="Year")
    
    return mergeDF
```
